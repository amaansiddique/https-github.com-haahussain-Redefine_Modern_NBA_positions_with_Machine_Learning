```python
from tqdm import tqdm
from bs4 import BeautifulSoup
import requests
import time
import pandas as pd
soup = BeautifulSoup
```

```python
def getAndParseUrl(url):
    results = requests.get(url)
    time.sleep(1)
    soup = BeautifulSoup(results.text, 'html.parser')
    return soup
```

```python
def get_playerz(): # function to grab every player and his stats from the page as a dictionary, and store as list
    
    # player dictionaries container
    players = []
    
    # loop through each page(for this website, each page is accessed by 'offset={some multiple of 100}')
    # last page is offset=2500
    for i in tqdm(range(0,8600,100)):
        url = f'https://www.basketball-reference.com/play-index/psl_finder.cgi?request=1&match=single&type=per_poss&per_minute_base=36&per_poss_base=100&lg_id=NBA&is_playoffs=N&year_min=2001&year_max=2019&franch_id=&season_start=1&season_end=-1&age_min=0&age_max=99&shoot_hand=&height_min=0&height_max=99&birth_country_is=Y&birth_country=&birth_state=&college_id=&draft_year=&is_active=&debut_yr_nba_start=&debut_yr_nba_end=&is_hof=&is_as=&as_comp=gt&as_val=0&award=&pos_is_g=Y&pos_is_gf=Y&pos_is_f=Y&pos_is_fg=Y&pos_is_fc=Y&pos_is_c=Y&pos_is_cf=Y&qual=&c1stat=per&c1comp=gt&c1val=&c2stat=off_rtg&c2comp=gt&c2val=&c3stat=def_rtg&c3comp=gt&c3val=&c4stat=mp&c4comp=gt&c4val=&c5stat=ws&c5comp=gt&c6mult=&c6stat=&order_by=season&order_by_asc=&offset={i}'
        soup = getAndParseUrl(url)
        
        # select each td tag under tr tag: each td tag in player_soup is a player's stat, all stored as a list
        player_soup = soup.select('tr td')
        
        # create slicing points to group the 31 elements(html tags) into single list(player)
        j=0
        k=34
        
        # there are a 100 players per page, 31 stats per player == 3100 elements in player_soup (except last page)
        # len(player_soup) gives the number of stats on each page
        # divide that by 31 to get number of players per page
        for i in range(int(len(player_soup)/34)):
            
            # each player container (player with stats)
            each_player = {}
            
            # p is a single player with stats sliced from player_soup
            p = player_soup[j:k]
            
            #loop through p to create dictionary of each player
            for stat in p:
                # set each stat equal to the tag text (i.e. pts_per_g = 15.4)
                each_player[stat.attrs['data-stat']] = stat.text
            
            # add each player dictionary to the players container
            players.append(each_player)
            
            # move slicer to next player
            j+=34
            k+=34
        
    return players
```

```python
def get_players2(url): # function to grab every player and his stats from the page as a dictionary, and store as list
    
    # player dictionaries container
    players = []
    
    # loop through each page(for this website, each page is accessed by 'offset={some multiple of 100}')
    # last page is offset=2500
    for i in tqdm(range(0,8900,100)):
        url = url+str(i)
        soup = getAndParseUrl(url)
        
        # select each td tag under tr tag: each td tag in player_soup is a player's stat, all stored as a list
        player_soup = soup.select('tr td')
        
        # create slicing points to group the 31 elements(html tags) into single list(player)
        j=0
        k=33
        
        # there are a 100 players per page, 31 stats per player == 3100 elements in player_soup (except last page)
        # len(player_soup) gives the number of stats on each page
        # divide that by 31 to get number of players per page
        for i in range(int(len(player_soup)/33)):
            
            # each player container (player with stats)
            each_player = {}
            
            # p is a single player with stats sliced from player_soup
            p = player_soup[j:k]
            
            #loop through p to create dictionary of each player
            for stat in p:
                # set each stat equal to the tag text (i.e. pts_per_g = 15.4)
                each_player[stat.attrs['data-stat']] = stat.text
            
            # add each player dictionary to the players container
            players.append(each_player)
            
            # move slicer to next player
            j+=33
            k+=33

        next_pg_checker = soup.select('p a[href]')
        if 'Next page' not in [c.text for c in next_pg_checker]:
            break
        else: continue
        
    return players
```

```python
def PERgreat_playerz(): # function to grab every player and his stats from the page as a dictionary, and store as list
    
    # player dictionaries container
    players = []
    
    # loop through each page(for this website, each page is accessed by 'offset={some multiple of 100}')
    # last page is offset=2500
    for i in tqdm(range(0,8700,100)):
        url = f'https://www.basketball-reference.com/play-index/psl_finder.cgi?request=1&match=single&type=per_poss&per_minute_base=36&per_poss_base=100&lg_id=NBA&is_playoffs=N&year_min=2001&year_max=2019&franch_id=&season_start=1&season_end=-1&age_min=0&age_max=99&shoot_hand=&height_min=0&height_max=99&birth_country_is=Y&birth_country=&birth_state=&college_id=&draft_year=&is_active=&debut_yr_nba_start=&debut_yr_nba_end=&is_hof=&is_as=&as_comp=gt&as_val=0&award=&pos_is_g=Y&pos_is_gf=Y&pos_is_f=Y&pos_is_fg=Y&pos_is_fc=Y&pos_is_c=Y&pos_is_cf=Y&qual=&c1stat=per&c1comp=gt&c1val=&c2stat=&c2comp=&c2val=&c3stat=&c3comp=&c3val=&c4stat=&c4comp=&c4val=&c5stat=&c5comp=&c6mult=&c6stat=&order_by=season&order_by_asc=&offset={i}'


        soup = getAndParseUrl(url)
        
        # select each td tag under tr tag: each td tag in player_soup is a player's stat, all stored as a list
        player_soup = soup.select('tr td')
        
        # create slicing points to group the 31 elements(html tags) into single list(player)
        j=0
        k=33
        
        # there are a 100 players per page, 31 stats per player == 3100 elements in player_soup (except last page)
        # len(player_soup) gives the number of stats on each page
        # divide that by 31 to get number of players per page
        for i in range(int(len(player_soup)/32)):
            
            # each player container (player with stats)
            each_player = {}
            
            # p is a single player with stats sliced from player_soup
            p = player_soup[j:k]
            
            #loop through p to create dictionary of each player
            for stat in p:
                # set each stat equal to the tag text (i.e. pts_per_g = 15.4)
                each_player[stat.attrs['data-stat']] = stat.text
            
            # add each player dictionary to the players container
            players.append(each_player)
            
            # move slicer to next player
            j+=33
            k+=33
        
    return players
```

```python
def get_playerz(): # function to grab every player and his stats from the page as a dictionary, and store as list
    
    # player dictionaries container
    players = []
    
    # loop through each page(for this website, each page is accessed by 'offset={some multiple of 100}')
    # last page is offset=2500
    for i in tqdm(range(0,8600,100)):
        url = f'https://www.basketball-reference.com/play-index/psl_finder.cgi?request=1&match=single&type=per_poss&per_minute_base=36&per_poss_base=100&lg_id=NBA&is_playoffs=N&year_min=2001&year_max=2019&franch_id=&season_start=1&season_end=-1&age_min=0&age_max=99&shoot_hand=&height_min=0&height_max=99&birth_country_is=Y&birth_country=&birth_state=&college_id=&draft_year=&is_active=&debut_yr_nba_start=&debut_yr_nba_end=&is_hof=&is_as=&as_comp=gt&as_val=0&award=&pos_is_g=Y&pos_is_gf=Y&pos_is_f=Y&pos_is_fg=Y&pos_is_fc=Y&pos_is_c=Y&pos_is_cf=Y&qual=&c1stat=per&c1comp=gt&c1val=&c2stat=&c2comp=&c2val=&c3stat=&c3comp=&c3val=&c4stat=&c4comp=&c4val=&c5stat=&c5comp=&c6mult=&c6stat=&order_by=season&order_by_asc=&offset={i}'

        soup = getAndParseUrl(url)
        
        # select each td tag under tr tag: each td tag in player_soup is a player's stat, all stored as a list
        player_soup = soup.select('tr td')
        
        # create slicing points to group the 31 elements(html tags) into single list(player)
        j=0
        k=33
        
        # there are a 100 players per page, 31 stats per player == 3100 elements in player_soup (except last page)
        # len(player_soup) gives the number of stats on each page
        # divide that by 31 to get number of players per page
        for i in range(int(len(player_soup)/33)):
            
            # each player container (player with stats)
            each_player = {}
            
            # p is a single player with stats sliced from player_soup
            p = player_soup[j:k]
            
            #loop through p to create dictionary of each player
            for stat in p:
                # set each stat equal to the tag text (i.e. pts_per_g = 15.4)
                each_player[stat.attrs['data-stat']] = stat.text
            
            # add each player dictionary to the players container
            players.append(each_player)
            
            # move slicer to next player
            j+=33
            k+=33
        
    return players
```
